```python
# Let's visualize the values based on rows
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
plt.figure(figsize=(20,10))
cmap=sns.cubehelix_palette(light=1, as_cmap=True, reverse=True) # to hold the color we are gonna 
sns.heatmap(df.isnull(), cmap=cmap)
# heat map is a function of sns libray
```

```python
# to change the data types of the dataset
df["Income"]=df["Income"].astype("int") # Although it's not required here i did to cover what all can be done in here.
```

```python
# Replace the missing values with a median value
df["Family"].fillna(df["Family"].median(), inplace=True)  #inplace=True to override the dataframe
```

```python
df.head(2)
```

```python
# To check the frequency of the variable
df["Family"].value_counts()
```

```python
# To find the max frequency 
df["Family"].value_counts().idxmax()
```

```python
# to replace nan with highest frquency number
df["Family"].replace(np.NaN, "1", inplace=True)
```

```python
df.head(2)
```

```python
# incase our dependent variable has missing values.we remove the rows
df.dropna(subset=["CreditCard"], axis=0, inplace=True)
```

```python
df.head(3)
```

```python
# to keep the indexs as usual after removing the rows
df.reset_index(drop=True, inplace=True)
```

```python
# Post treating missing values
plt.figure(figsize=(20,10))
cmap=sns.cubehelix_palette(light=1, as_cmap=True, reverse=True) # to hold the color we are gonna 
sns.heatmap(df.isnull(), cmap=cmap)
```
